```python
import pandas as pd
import numpy as np
```

```python
prod = pd.read_csv('./data/PRODCL2.csv')
prod
```

```python
prod1 = prod[['소분류명','소비재분류']]
prod1.columns =['품목명','소비재분류']
```

```python
g1 = pd.read_csv('./data/1번군집.csv',encoding='cp949')
g1 = g1.set_index('고객번호')
g2 = pd.read_csv('./data/2번군집.csv',encoding='cp949')
g2 = g2.set_index('고객번호')
g3 = pd.read_csv('./data/3번군집.csv',encoding='cp949')
g3 = g3.set_index('고객번호')
```

```python
prod_col1 = prod.loc[prod['소비재분류']=='편의품']['소분류명'].values
g1_col = np.array(g1.columns.to_list()) 
ili1 = np.where(np.isin(g1_col, prod_col1)==True)[0]
g1 = g1[g1_col[ili1]]
g1
```

```python
prod_col2 = prod.loc[(prod['소비재분류']=='전문품') | (prod['소비재분류']=='선매품')]['소분류명'].values
g2_col = np.array(g2.columns.to_list()) 
ili2 = np.where(np.isin(g2_col, prod_col2)==True)[0]
g2 = g2[g2_col[ili2]]
g2
```

```python
prod_col3 = prod.loc[prod['소비재분류']=='선매품']['소분류명'].values
g3_col = np.array(g3.columns.to_list()) 
ili3 = np.where(np.isin(g3_col, prod_col3)==True)[0]
g3 = g3[g3_col[ili3]]
g3
```

```python
g12 = pd.concat([g1,g2])
ga = pd.concat([g12,g3])
ga.fillna(0,inplace=True)
ga
```

```python
# 코사인 유사도를 하기 위해서 전치 행렬을 구하는 단계 
ga_T = ga.transpose()
ga_T
```

```python
for a in ga.columns:
    print(a)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ga_T,ga_T)
# 코사인 유사도 만든 행렬을 데이터프레임으로 변환 
item_sim_df = pd.DataFrame(data=item_sim, index = ga.columns,
                          columns=ga.columns)
print(item_sim_df.shape)
item_sim_df.head(3)
```

```python
# 코사인유사도 내적곱한값을 코사인 유사도 절대값의 합으로 나눈 거(정규화를위해서)
import numpy as np
def predict_ga(ga_arr, item_sim_arr):
    ga_pred = ga_arr.dot(item_sim_arr)/np.array([np.abs(item_sim_arr).sum(axis=1)])
    return ga_pred
```

```python
# 개인별 상품 예측 
ga_pred = predict_ga(ga.values, item_sim_df.values)
ga_pred_matrix = pd.DataFrame(data=ga_pred, index = ga.index,\
                            columns = ga.columns)
ga_pred_matrix
```

```python
from sklearn.metrics import mean_squared_error

# mse =실제와 예측 점수차이 비교 
def get_mse(pred, actual):
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred,actual)

print('아이템 기반 모든 최근접 이웃 MSE:', get_mse(ga_pred, ga.values))
    
```

```python
def predict_ga_topsim(ga_arr, item_sim_arr, n =20):
    pred = np.zeros(ga_arr.shape)
    
    for col in range(ga_arr.shape[1]):
        top_n_items = [np.argsort(item_sim_arr[:,col])[:-n-1:-1]]
        for row in range(ga_arr.shape[0]):
            pred[row, col] = item_sim_arr[col,:][top_n_items].dot(ga_arr[row,:][top_n_items].T)
            
            pred[row, col] /= np.sum(np.abs(item_sim_arr[col,:][top_n_items]))
            
    return pred
        
```

```python

ga_pred = predict_ga_topsim(ga.values, item_sim_df.values, n=20)
print('아이템 기반 최근접 TOP-20 이웃 MSE:', get_mse(ga_pred, ga.values))

ga_pred_matrix = pd.DataFrame(data=ga_pred, index = ga.index, columns=ga.columns)
```

```python
def get_notbuy_product(ga,고객번호):
    user_ga = ga.loc[고객번호,:]

    already_buy = user_ga[user_ga > 0].index.tolist()

    ga3_list = ga.columns.tolist()

    notbuy_list = [pro for pro in ga3_list if pro not in already_buy]
    
    return notbuy_list
```
