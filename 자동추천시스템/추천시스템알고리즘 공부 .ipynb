```python
import pandas as pd
import numpy

p= pd.read_csv('test2.csv',encoding='cp949')

p = p.set_index('고객번호')
p
```

```python
p_T = p.transpose()
p_T.head(3)
```

```python
### 코사인 유사도 
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(p_T,p_T)

item_sim_df = pd.DataFrame(data=item_sim, index = p.columns,
                          columns=p.columns)
print(item_sim_df.shape)
item_sim_df.head(3)
```

```python
item_sim_df["14K"].sort_values(ascending=False)[:6]
```

```python
import numpy as np
def predict_p(p_arr, item_sim_arr):
    p_pred = p_arr.dot(item_sim_arr)/np.array([np.abs(item_sim_arr).sum(axis=1)])
    return p_pred
```

```python
p_pred = predict_p(p.values, item_sim_df.values)
p_pred_matrix = pd.DataFrame(data=p_pred, index = p.index,\
                            columns = p.columns)
p_pred_matrix.head(3)
```

```python
from sklearn.metrics import mean_squared_error


def get_mse(pred, actual):
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred,actual)

print('아이템 기반 모든 최근접 이웃 MSE:', get_mse(p_pred, p.values))
    
```

```python
def predict_p_topsim(p_arr, item_sim_arr, n =20):
    pred = np.zeros(p_arr.shape)
    
    for col in range(p_arr.shape[1]):
        top_n_items = [np.argsort(item_sim_arr[:,col])[:-n-1:-1]]
        for row in range(p_arr.shape[0]):
            pred[row, col] = item_sim_arr[col,:][top_n_items].dot(p_arr[row,:][top_n_items].T)
            
            pred[row, col] /= np.sum(np.abs(item_sim_arr[col,:][top_n_items]))
            
    return pred
```

```python
p_pred = predict_p_topsim(p.values, item_sim_df.values, n=20)
print('아이템 기반 최근접 TOP-20 이웃 MSE:', get_mse(p_pred, p.values))

p_pred_matrix = pd.DataFrame(data=p_pred, index = p.index, columns=p.columns)
```

```python
def get_unseen_movies(p,s):

    user_p = p.loc[s,:]

    already_seen = user_p[user_p > 0].index.tolist()

    movies_list = p.columns.tolist()
    
    unseen_list = [movie for movie in movies_list if movie not in already_seen]
    
    return unseen_list
```

```python
def recomm_movie_by_userid(pred_df, s, unseen_list, top_n=5):
    recomm_movies = pred_df.loc[s, unseen_list].sort_values(ascending=False)[:top_n]
    return recomm_movies

unseen_list = get_unseen_movies(p,500)

recomm_movies = recomm_movie_by_userid(p_pred_matrix, 500, unseen_list, top_n=5)

recomm_movies = pd.DataFrame(data=recomm_movies.values, index = recomm_movies.index,
                            columns=['pred_score'])

recomm_movies 
```
