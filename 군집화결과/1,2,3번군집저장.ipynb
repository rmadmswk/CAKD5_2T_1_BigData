```python
import pandas as pd
import numpy as np

df = pd.read_csv('라벨붙임.csv',encoding='cp949')
df = df.drop('Unnamed: 0',axis=1)
df
```

```python
df1 = df[df['kmeans_label']==0] # 1번군집
df2 = df[df['kmeans_label']==1] # 2번군집
df3 = df[df['kmeans_label']==2] # 3번군집
df1 = df1[['고객번호']]
df2 = df2[['고객번호']]
df3 = df3[['고객번호']]
```

```python
# 소분류,고객번호 테이블을 불러옴
a = pd.read_csv('test21.csv',encoding='cp949')
```

```python
# 1번군집 + 소분류
b = pd.merge(df1,a,on='고객번호',how='left')
```

```python
# 2번군집 + 소분류
c = pd.merge(df2,a,on='고객번호',how='left')
```

```python
# 3번군집 + 소분류
d = pd.merge(df3,a,on='고객번호',how='left')
```

```python
import csv
b.to_csv('1번군집.csv',mode="w",encoding='euc-kr',index=False)
```

```python
import csv
c.to_csv('2번군집.csv',mode="w",encoding='euc-kr',index=False)
```

```python
import csv
d.to_csv('3번군집.csv',mode="w",encoding='euc-kr',index=False)
```
