```python
import pandas as pd
import excel
yedf=pd.read_excel('data/분기별.xlsx')
yedf
```

```python
pyedf = pd.pivot_table(yedf,
                     index = '고객번호', # 행 위치에 들어갈 열
                     columns = '분기', # 열 위치에 들어갈 열
                     values = ['구매횟수','총구매액','구매평균'], # 데이터로 사용할 열
                     aggfunc = 'max') # 데이터 집계 함수
dd=pyedf['총구매액']
dd
```

```python
dd
import numpy as np
dd['14_1구매경험'] = np.where(dd['14_1']>0,'유','무')
dd['14_2구매경험'] = np.where(dd['14_2']>0,'유','무')
dd['14_3구매경험'] = np.where(dd['14_3']>0,'유','무')
dd['14_4구매경험'] = np.where(dd['14_4']>0,'유','무')
dd['15_1구매경험'] = np.where(dd['15_1']>0,'유','무')
dd['15_2구매경험'] = np.where(dd['15_2']>0,'유','무')
dd['15_3구매경험'] = np.where(dd['15_3']>0,'유','무')
dd['15_4구매경험'] = np.where(dd['15_4']>0,'유','무')
dd
```

```python
dd=dd[(dd['14_1구매경험']=='유') & (dd['14_2구매경험']=='유')
       & (dd['14_3구매경험']=='유') & (dd['14_4구매경험']=='유')
       & (dd['15_1구매경험']=='유') & (dd['15_2구매경험']=='유')
       & (dd['15_3구매경험']=='유') & (dd['15_4구매경험']=='유')]

dd
# dd.columns
# dd[['고객번호','14_1','14_2','14_3','14_4','15_1','15_2','15_3','15_4']]
```

```python
dd.fillna(0)
da=dd.copy()
```

```python
# aa=pd.qcut(dd[['14_1']].sort_values(by='14_1'),10)
import numpy as np

aa=da['14_1'].sort_values(ascending=False)
cuts = pd.qcut(aa,10,labels=np.arange(1,11)[::-1])
cc=pd.concat([aa,cuts],axis=1)
cc

```

```python
lst=['14_2','14_3','14_4','15_1','15_2','15_3','15_4']
for i in lst:
    aa=da[f'{i}'].sort_values(ascending=False)
    
    cuts = pd.qcut(aa,10,labels=np.arange(1,11)[::-1])
    dd=pd.concat([aa,cuts],axis=1)  
    cc=pd.merge(cc,dd, on='고객번호',how='left')
    display(cc)
```
