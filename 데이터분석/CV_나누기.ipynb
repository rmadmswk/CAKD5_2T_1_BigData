```python
import pandas as pd
import numpy as np

df = pd.read_csv('./dataset/MainCategory/편의품.csv')
df_com = df.drop('Unnamed: 0',axis=1)
df_com.head()
```

```python
def auto(df1,text):
    
    def add_CV(df,text):
        display(df)
        for i in df.index:
            m=df.iloc[i,1:].mean()
            if m == 0:
                cv=0
            elif m != 0:
                cv = (df.iloc[i,1:].std())/m
            df.loc[i,text+'CV'] = cv
        df_r=df[['고객번호',text+'CV']]
        return df_r
    
    def cut_seals(df1,text):
        c=df1.columns[:8]
        df1=df1[c]
        df1_new=add_CV(df1,text)
        return df1_new
    
    def cut_counts(df1,text):
        c=['고객번호']
        c.extend(df1.columns[9:-1])
        df1=df1[c]
        df1_new=add_CV(df1,text)
        return df1_new

    display(df1)
    df3=cut_seals(df1,text)
    df4=cut_counts(df1,text)
    df3.columns=['고객번호',text+'CV_S']
    df4.columns=['고객번호',text+'CV_C']
    display(df3)
    display(df4)
    df_sum=pd.merge(df3,df4,on='고객번호')
    return df_sum
```

```python
# df_new=auto(df_com,'편의품')
df_new.columns = ['고객번호','편의품_CV_S','편의품_CV_C']
df_new
```
