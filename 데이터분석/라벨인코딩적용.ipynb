```python
import pandas as pd
import cx_Oracle
import csv
import numpy as np
a = pd.read_csv('sample_re.csv')
a
```

```python
a.drop(['고객번호.1'],axis=1,inplace=True)
```

```python
a
```

```python
from sklearn.preprocessing import LabelEncoder
depts = ['하위','보통','상위','최상위']

# LabelEncoder를 객체로 생성
encoder = LabelEncoder()

# fit, transform 메소드를 통한 레이블 인코딩
encoder.fit(depts)
labels = encoder.transform(depts)

print('레이블 인코딩 클래스 : ',encoder.classes_) # 인코딩 클래스는 classes_속성값으로 확인
print('레이블 인코딩 변환값 : ',labels)
```

```python
a['그룹'] = encoder.fit_transform(a['그룹'])
a
```

```python
a.to_csv('지역라벨인코딩.csv')
```
