```python
def auto(df,df1,text):
    def div_col(df):
        c=['고객번호']
        c.extend(df.columns[8:])
        df1 = df[c]
        return df1
    
    def add_CV(df,text):
        for i in df.index:
            m=df.iloc[i,1:].mean()
            if m == 0:
                cv=0
            elif m != 0:
                cv = (df.iloc[i,1:].std())/m
            df.loc[i,text+'CV'] = cv
        df_r=df[['고객번호',text+'CV']]
        return df_r
    
    def cut_seals(df1,text):
        c=df1.columns[:8]
        df1=df1[c]
        df1_new=add_CV(df1,text)
        return df1_new

    df2=div_col(df)
    display(df2)
    df3=cut_seals(df1,text)
    display(df3)
    df_sum=pd.merge(df2,df3,on='고객번호')
    return df_sum

df_com_CV=auto(cc2,cc3,'1기객단가')
display(df_com_CV)
```
