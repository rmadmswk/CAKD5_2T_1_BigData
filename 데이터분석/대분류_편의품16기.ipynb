```python
import pandas as pd
import numpy as np
```

```python
b=pd.read_csv('대분류세분류테이블.csv',encoding='cp949')
```

```python
a=pd.read_csv('16기존고객.csv',encoding='cp949')
```

```python
c=pd.read_csv('17기존고객.csv',encoding='cp949')
```

```python
df_f=b[(b['소비재분류']=='편의품')]
df_f
```

```python
df_pivot_f = pd.pivot_table(df_f,                # 피벗할 데이터프레임
                     index = '고객번호',    # 행 위치에 들어갈 열
                     columns = 'QUARTER',    # 열 위치에 들어갈 열
                     values = '영수증번호',     # 데이터로 사용할 열
                     aggfunc = 'count')   # 데이터 집계함수
```

```python
df_pivot_f = df_pivot_f.fillna(0)
```

```python
# 7기,8기 드랍 테이블
df_pivot_f2 = df_pivot_f.drop(['7Q','8Q'],axis=1)
```

```python
df_pivot_f2['편의품구매횟수'] = df_pivot_f2['1Q'] + df_pivot_f2['2Q'] + df_pivot_f2['3Q'] + df_pivot_f2['4Q'] + df_pivot_f2['5Q']+ df_pivot_f2['6Q']
```

```python
df_pivot_f3 = df_pivot_f2.drop(['1Q','2Q','3Q','4Q','5Q','6Q'],axis=1)
```

```python
# 6기 구매횟수
tree=pd.merge(a, df_pivot_f3,on='고객번호', how='left')
tree
```

```python
# 7기,8기 드랍 테이블
df_pivot_f4 = df_pivot_f.drop(['8Q'],axis=1)
```

```python
df_pivot_f4['편의품구매횟수'] = df_pivot_f4['1Q'] + df_pivot_f4['2Q'] + df_pivot_f4['3Q'] + df_pivot_f4['4Q'] + df_pivot_f4['5Q']+ df_pivot_f4['6Q'] +df_pivot_f4['6Q']
```

```python
df_pivot_f5 = df_pivot_f4.drop(['1Q','2Q','3Q','4Q','5Q','6Q','7Q'],axis=1)
```

```python
# 7기 구매횟수
tree3=pd.merge(c, df_pivot_f5,on='고객번호', how='left')
tree3
```

```python
df_pivot_s4 = pd.pivot_table(df_f,                # 피벗할 데이터프레임
                     index = '고객번호',    # 행 위치에 들어갈 열
                     columns = 'QUARTER',    # 열 위치에 들어갈 열
                     values = '구매금액',     # 데이터로 사용할 열
                     aggfunc = 'sum')   # 데이터 집계함수
```

```python
df_pivot_s5 = df_pivot_s4.drop(['8Q','7Q'],axis=1)
```

```python
aa=df_pivot_s5['1Q'].sort_values(ascending=False)
cuts = pd.qcut(aa,10,labels=np.arange(1,11)[::-1])
cc=pd.concat([aa,cuts],axis=1)
```

```python
lst=['2Q','3Q','4Q','5Q','6Q']
for i in lst:
    aa=df_pivot_s5[f'{i}'].sort_values(ascending=False)
    
    cuts = pd.qcut(aa,10,labels=np.arange(1,11)[::-1])
    table=pd.concat([aa,cuts],axis=1)  
    cc=pd.merge(cc,table, on='고객번호',how='left')
```

```python
cc.columns=['편의품1','편의품등급1','편의품2','편의품등급2','편의품3','편의품등급3','편의품4','편의품등급4','편의품5','편의품등급5','편의품6','편의품등급6']
```

```python
cc = cc.drop(['편의품1','편의품2','편의품3','편의품4','편의품5','편의품6'],axis=1)
```

```python
cc = cc.reset_index()
```

```python
cc = cc.sort_values('고객번호')
```

```python
tree2=pd.merge(a, cc,on='고객번호', how='left')
tree2
```
